def categorize_customers(df: pd.DataFrame) -> pd.DataFrame:
    """
    Categorize customers based on their order numbers.
    
    Args:
        df (pd.DataFrame): DataFrame containing customer order data with 'order_number' column
    
    Returns:
        pd.DataFrame: DataFrame with added 'label' column categorizing customers
    """
    # Create a copy to avoid modifying the original dataframe
    result_df = df.copy()
    
    # Define customer categories based on order numbers
    result_df.loc[result_df['order_number'] > 40, 'label'] = 'loyal customer'
    result_df.loc[(result_df['order_number'] > 10) & 
                 (result_df['order_number'] <= 40), 'label'] = 'Regular customer'
    result_df.loc[result_df['order_number'] <= 10, 'label'] = 'New customer'
    
    return result_df

# Apply the categorization
df = categorize_customers(df)

# Aggregating Data with transform()

```python

def get_max_value(series):
    """
    Returns the maximum value in a pandas Series.
    
    Args:
        series (pd.Series): The input series to find maximum value from
        
    Returns:
        float or int: The maximum value in the series
    """
    return series.max()

# Add a column with the maximum order number for each user
ords_prods_merge_peaks['max_order'] = (
    ords_prods_merge_peaks
    .groupby(['user_id'])['order_number']
    .transform(get_max_value)
)
```

# The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:
If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
#Performing a Single Aggregation for prices by user_id
user_avg_prices = ords_prods_merge_peaks.groupby('user_id').agg({'prices': 'mean'})['prices']

# adding column 
ords_prods_merge_peaks['avg_value'] = ords_prods_merge_peaks['user_id'].map(user_avg_prices)

ords_prods_merge_peaks.head()
```

```python
#applying the logic to identify high and low spenders
ords_prods_merge_peaks.loc[ords_prods_merge_peaks['avg_value'] < 10, 'spenders_flag'] = 'Low spender'
ords_prods_merge_peaks.loc[ords_prods_merge_peaks['avg_value'] >= 10, 'spenders_flag'] = 'High spender'
# Get the count of each unique value in the 'Loyalty flag' column, including NaN values
ords_prods_merge_peaks['spenders_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge_peaks.groupby(['spenders_flag']).user_id.nunique()
```

```python
#returning specific colomns to check the result
ords_prods_merge_peaks[['user_id', 'loyalty_flag', 'spenders_flag']].head(50)
```

```python
#Performing a Single Aggregation for days_since_prior_order by user_id
frequency = ords_prods_merge_peaks.groupby('user_id').agg({'days_since_prior_order': 'mean'})['days_since_prior_order']

# adding column 
ords_prods_merge_peaks['frequency'] = ords_prods_merge_peaks['user_id'].map(frequency)

ords_prods_merge_peaks.head()
```
