```python
user_1 = df_ords_renamed[df_ords_renamed['user_id'] == '1']
df_prods[df_prods['department_id']==14]
print(user_1)

```

# You also need to provide some details about this user’s behavior. What basic stats can you provide based on the information you have?

```python
# Total number of orders
total_orders = user_1.shape[0]

# Average days since previous order (excluding first NaN)
avg_days_between_orders = user_1['days_since_prior_order'].mean()

# Most frequent hour of day for orders
most_common_hour = user_1['hour'].mode()[0]

# Most frequent day of week for orders
most_common_day = user_1['orders_day_of_week'].mode()[0]

print(f"Total orders placed: {total_orders}")
print(f"Average days between orders: {avg_days_between_orders:.2f}")
print(f"Most common order hour: {most_common_hour}")
print(f"Most common order day of week: {most_common_day}")
```
